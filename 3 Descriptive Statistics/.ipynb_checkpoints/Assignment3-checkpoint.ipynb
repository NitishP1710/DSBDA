```python
'''
Descriptive Statistics - Measures of Central Tendency and variability

Perform the following operations on any open-source dataset (e.g., data.csv)
1. Provide summary statistics (mean, median, minimum, maximum, standard
deviation) for a dataset (age, income etc.) with numeric variables grouped by
one of the qualitative (categorical) variable. For example, if your categorical
variable is age groups and quantitative variable is income, then provide
summary statistics of income grouped by the age groups. Create a list that
contains a numeric value for each response to the categorical variable.
2. Write a Python program to display some basic statistical details like percentile,
mean, standard deviation etc. of the species of ‘Iris-setosa’, ‘Iris-versicolor’
and ‘Iris- verginica’ of iris.csv dataset.
Provide the codes with outputs and explain everything that you do in this step.
'''
```

```python
import pandas as pd;
import numpy as np
```

```python
df=pd.read_csv("Iris.csv")
print(df)
```

```python
print(df.dtypes)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
stats=df.groupby("Species").agg(['mean', 'median', 'min', 'max', 'std'])
print(stats)
```

```python
#create a list with mean value for each species
sepal_length_mean = df.groupby('Species')['SepalLengthCm'].mean().tolist()
sepal_length_mean
```

```python
#get statistical data using describe() and percentiles
species_stats = df.groupby('Species').describe(percentiles=[0.25,0.5,0.75])
species_stats
```

```python
#get mean values for species
print("Mean value of petal length for Iris virginica: ",df[df['Species']=='Iris-virginica']['PetalLengthCm'].mean())
```

```python
#get specific percentile values for species
print("25th percentile value of petal width for Iris versicolor: ",df[df['Species']=='Iris-versicolor']['PetalWidthCm'].quantile(0.25))
```

```python
#get standard deviation values for species
print("std value of Sepal length for Iris setosa: ",df[df['Species']=='Iris-setosa']['SepalLengthCm'].std())
```
