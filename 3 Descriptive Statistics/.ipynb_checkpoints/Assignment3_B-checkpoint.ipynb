```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
print(df)
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
#show stats using groupby
stats = df.groupby('AgeGroup').agg(['mean','median','std','min','max'])
stats
```

```python
#get list of median age for each group
age_group_median = df.groupby('AgeGroup')['Age'].median().tolist()
age_group_median
```

```python
#get statistical data using describe() and percentiles
ageGrp_stats = df.groupby('AgeGroup').describe(percentiles=[0.25,0.5,0.75])
ageGrp_stats
```

```python
#get mean value for income
print("Mean value of Income for age group 26-35 : ",df[df['AgeGroup']=='26-35']['Income'].mean())
```

```python
#get median value for income
print("Median value of Income for age group 36-45 : ",df[df['AgeGroup']=='36-45']['Income'].median())
```

```python
#get 75th percentile for income
print("value of 75th percentile Income for age group 18-25 : ",df[df['AgeGroup']=='18-25']['Income'].quantile(0.75))
```
