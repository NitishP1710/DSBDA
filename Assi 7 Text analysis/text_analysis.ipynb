## Term frequency and inverse Document frequency


```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer

corpus = [
    'the cat sat on the mat',
    'the dog sat on the log'
]

# --- PART 1: Term Frequency (TF) ---
# CountVectorizer counts how many times each word appears
cv = CountVectorizer()
tf_matrix = cv.fit_transform(corpus)
words = cv.get_feature_names_out()

df_tf = pd.DataFrame(tf_matrix.toarray(), columns=words)
print("--- Term Frequency (Raw Counts) ---")
print(df_tf)
print('='*45)
row_sum=df_tf.sum(axis=1)
normalized_dftf=df_tf.div(row_sum,axis=0)
print("--- Normalized Term Frequency (TF) ---")
print(normalized_dftf)

# --- PART 2: Inverse Document Frequency (IDF) ---
# TfidfTransformer calculates the weights of the words across the corpus
tfidf_trans = TfidfTransformer()
tfidf_trans.fit(tf_matrix)

df_idf = pd.DataFrame(tfidf_trans.idf_, index=words, columns=["IDF_Weight"])
print("\n--- Inverse Document Frequency (Weights) ---")
print(df_idf.sort_values(by="IDF_Weight", ascending=False))
```

```python
import math

docs = [["cat", "hat"], ["cat", "dog"]]
all_words = ["cat", "hat", "dog"]

# 1. Manual IDF
num_docs = len(docs)
idf_values = {}
for word in all_words:
    count = sum(1 for d in docs if word in d)
    idf_values[word] = math.log10(num_docs / count)

print("IDF Values:", idf_values)

# 2. Manual TF-IDF for Doc 1
doc1 = docs[0]
for word in all_words:
    tf = doc1.count(word) / len(doc1)
    tfidf = tf * idf_values[word]
    print(f"Word: {word} | TF: {tf} | IDF: {idf_values[word]:.4f} | TF-IDF: {tfidf:.4f}")
```
