```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer
from nltk.stem import PorterStemmer,WordNetLemmatizer
```

```python
file1=open('file1.txt')
file2=open('file2.txt')
text1=file1.read()
text2=file2.read()
print(text1)
print('*'*100)
print(text2)
```

```python
token1=word_tokenize(text1)
print(token1)
print(len(token1))
print('*'*100)
token2=word_tokenize(text2)
print(token2)
print(len(token2))
```

```python
stopwords=set(stopwords.words('english'))
filtered_token1=[word for word in token1 if word not in stopwords]
print(filtered_token1)
print(len(filtered_token1))
print('*'*100)
filtered_token2=[word2 for word2 in token2 if word2 not in stopwords]
print(filtered_token2)
print(len(filtered_token2))
```

```python
pos_tag1=pos_tag(filtered_token1)
print("Pos tag for text 1")
for word,pos_tagg in pos_tag1:
    print(f"{word} : {pos_tagg}")

pos_tag2=pos_tag(filtered_token2)
print("Pos tag for text 2")
for word,pos_tagg in pos_tag2:
    print(f"{word} : {pos_tagg}")
```

```python
stemmer=PorterStemmer()
stemmed_word1=[stemmer.stem(word) for word in filtered_token1]
print(stemmed_word1)
print('*'*100)

stemmed_word2=[stemmer.stem(word) for word in filtered_token2]
print(stemmed_word2)
```

```python
lemmatizer=WordNetLemmatizer()
lemmatized_word1=[lemmatizer.lemmatize(word) for word in filtered_token1]
print(lemmatized_word1)
print('*'*100)
lemmatized_word2=[lemmatizer.lemmatize(word) for word in filtered_token2]
print(lemmatized_word2)
```
