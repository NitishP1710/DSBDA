```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("IRIS.csv")
print(df.head)
```

```python
stats=df.groupby("species").agg(['mean', 'median', 'min', 'max', 'std'])
print(stats)
```

```python
print(df)
```

```python
#box plot iris vs sepal length

sns.boxplot(data=df,x='species',y='sepal_length')
plt.title("species vs sepal_length")
plt.show()
#Sepal Length data is positively skewed median value is between 5.5-6 No outliers present
```

```python
#Boxplot species vs sepal_width
sns.boxplot(data=df,x='species',y='sepal_width')
plt.title("species vs sepal_width")
plt.show()

#Sepal Width Data is not much skewed median value is close to 3 It has outliers

```

```python
#Boxplot species vs petal_length
sns.boxplot(data=df,x='species',y='petal_length')
plt.title("species vs petal_length")
plt.show()
#Petal Length Data is not uniform Data is dispersed Positively Skewed Median value is between 4-5 No outliers

```

```python
#Boxplot species vs  petal_width 
sns.boxplot(data=df,x='species',y='petal_width')
plt.title("species vs  petal_width")
plt.show()
#Petal Width Data is not uniform Data is dispersed Positively Skewed Median value is between 1-1.5 No Outliers
```

```python
sns.histplot(df['sepal_length'], kde=True, bins=30)
plt.title("Distribution of sepal_length")
plt.xlabel("sepal_length")
plt.ylabel("Frequency")
plt.show()
#Sepal Length data is positively skewed mode value is near 5
```

```python
sns.histplot(df['sepal_width'], kde=True, bins=30)
plt.title("Distribution of sepal_width")
plt.xlabel("sepal_width")
plt.ylabel("Frequency")
plt.show()

#Sepal Width Data is not much skewed mean,mode,median are close to 3
```

```python
sns.histplot(df['petal_length'], kde=True, bins=30)
plt.title("Distribution of petal_length")
plt.xlabel("petal_length")
plt.ylabel("Frequency")
plt.show()
#Petal Length Data is not uniform It is multimodal (multiple peaks) Mode value is between 1-2
```

```python
sns.histplot(df['petal_width'], kde=True, bins=30)
plt.title("Distribution of petal_width")
plt.xlabel("petal_width")
plt.ylabel("Frequency")
plt.show()
#Petal Width Data is not uniform It is multimodal (multiple peaks) Mode value is between 0-0.5
```
