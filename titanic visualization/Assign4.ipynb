```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv("Titanic.csv")
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df=df.drop("Cabin",axis=1)
```

```python
#fill null values mean for age and mode for Embarked
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
#plot of count of survived and not survived passenger
sns.countplot(data=df, x='Survived')
plt.title("Survival Count")
plt.show()

#inferrence: Number of death are more than number of survivied
```

```python
#plot of count of survived and not survived passenger gender wise
sns.countplot(data=df, x='Survived',hue='Sex')
plt.title("Gender wise Survival Count")
plt.show()

#inference : 1)Number of male death are more than number of female death 2)Number of female survival more than male
#conclusion : First priority may be given to female for rescue
```

```python

#plot of count of survived and not survived passenger by embarked
sns.countplot(data=df, x='Age',hue='Survived')
plt.title("Survival Count by Age")
plt.show()
```

```python
sns.boxplot(data=df,x='Survived',y='Age')
plt.title("Age vs Survival")
plt.show()

#the median of death is in 25 to 35 range, Most of the old (Outlieries unable to survive)
```

```python
sns.scatterplot(data=df, x='Age', y='Fare')
plt.title('Age vs Fare')
plt.show()
```

```python
sns.histplot(df['Fare'], kde=True, bins=20)
plt.title("Distribution of Ticket Fare")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()

#inference : Positively skewed,As fare increases the number of pa
```

```python
sns.lineplot(data=df, x='Age', y='Fare', hue='Survived', marker='o')
plt.title("Average Fare vs Age for Survivors and Non-Survivors")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
sns.boxplot(data=df,x='Sex',y='Age',hue="Survived")
plt.title("Gender vs Survival")
plt.show()
```
