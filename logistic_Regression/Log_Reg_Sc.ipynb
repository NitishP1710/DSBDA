```python
import pandas as pd
import numpy as np
```

```python
# 1. Load the dataset
df = pd.read_csv("data.csv")
```

```python
# 2. Preprocessing: Convert Gender to numbers and create numpy arrays
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})
X = df[['Age', 'EstimatedSalary', 'Gender']].values
y = df['Purchased'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# 4. Feature Standardization from scratch (z-score normalization)
# Formula: (X - mean) / standard_deviation
def standardize(X_train, X_test):
    mean = np.mean(X_train, axis=0)
    std = np.std(X_train, axis=0)
    X_train_scaled = (X_train - mean) / std
    X_test_scaled = (X_test - mean) / std
    return X_train_scaled, X_test_scaled

X_train_scaled, X_test_scaled = standardize(X_train, X_test)
```

```python
# 5. Logistic Regression Model from Scratch
class LogisticRegressionScratch:
    def __init__(self, learning_rate=0.1, iterations=2000):
        self.lr = learning_rate
        self.iterations = iterations
        self.weights = None
        self.bias = None

    def sigmoid(self, z):
        # Clip z to prevent exponential overflow
        z = np.clip(z, -250, 250)
        return 1 / (1 + np.exp(-z))

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Gradient Descent Optimization
        for _ in range(self.iterations):
            # linear model: z = w.X + b
            linear_model = np.dot(X, self.weights) + self.bias
            # y_pred = sigmoid(z)
            y_predicted = self.sigmoid(linear_model)

            # Calculate gradients
            dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y))
            db = (1 / n_samples) * np.sum(y_predicted - y)

            # Update weights & bias
            self.weights -= self.lr * dw
            self.bias -= self.lr * db

    def predict(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        y_predicted = self.sigmoid(linear_model)
        # Convert probabilities to binary classes (Threshold = 0.5)
        y_predicted_cls = [1 if p > 0.5 else 0 for p in y_predicted]
        return np.array(y_predicted_cls)
```

```python
# Train the model
model = LogisticRegressionScratch(learning_rate=0.1, iterations=2000)
model.fit(X_train_scaled, y_train)
```

```python
# Make Predictions
y_pred = model.predict(X_test_scaled)
```

```python
# 6. Calculate Confusion Matrix Elements from Scratch
# True Positives: Predicted 1, Actual 1
TP = np.sum((y_test == 1) & (y_pred == 1))
# False Positives: Predicted 1, Actual 0
FP = np.sum((y_test == 0) & (y_pred == 1))
# True Negatives: Predicted 0, Actual 0
TN = np.sum((y_test == 0) & (y_pred == 0))
# False Negatives: Predicted 0, Actual 1
FN = np.sum((y_test == 1) & (y_pred == 0))

# 7. Calculate Metrics Using Mathematical Formulas
total_samples = len(y_test)
accuracy = (TP + TN) / total_samples
error_rate = (FP + FN) / total_samples  # Or simply: 1 - accuracy
precision = TP / (TP + FP) if (TP + FP) > 0 else 0
recall = TP / (TP + FN) if (TP + FN) > 0 else 0
```

```python
# 8. Display Results
print("--- Confusion Matrix Components ---")
print(f"True Positives (TP): {TP}")
print(f"False Positives (FP): {FP}")
print(f"True Negatives (TN): {TN}")
print(f"False Negatives (FN): {FN}")

print("\n--- Performance Metrics ---")
print(f"Accuracy: {accuracy}")
print(f"Error Rate: {error_rate}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
```
