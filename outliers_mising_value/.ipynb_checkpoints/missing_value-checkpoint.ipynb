```python
import pandas as pd
```

```python
df=pd.read_csv("missingvalues.csv")
```

```python
print(df.head())
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
#df['math score']=df['math score'].astype('float64') : this will give value error for nan as nan have datatype string 
df['math score'] = pd.to_numeric(df['math score'], errors='coerce')
```

```python
df.dtypes
```

```python
df['reading score'] = pd.to_numeric(df['reading score'], errors='coerce')
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df['math score']=df['math score'].fillna(df['math score'].mean())
df['reading score']=df['reading score'].fillna(df['reading score'].mean())
df['writing score']=df['writing score'].fillna(df['writing score'].mean())
df['Placement Score']=df['Placement Score'].fillna(df['Placement Score'].mean())
df['Region']=df['Region'].fillna(df['Region'].mode()[0])
```
